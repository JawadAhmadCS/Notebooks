```python
import numpy as np
import matplotlib.pyplot as plt
from stingray import Lightcurve
```

```python
time = np.linspace(0, 100, 1000)  # Simulated time in seconds
counts = np.random.poisson(100, 1000)  # Simulated photon counts

lc = Lightcurve(time, counts)  # Create a light curve
lc.plot()  # Plot the light curve
```

```python
# Sample data
time = np.arange(0, 100, 0.1)
counts = np.random.normal(0, 1, len(time))

# Create a Lightcurve object
lc = Lightcurve(time, counts)

# Manually normalize the lightcurve (normalize by max value)
normalized_counts = lc.counts / np.max(lc.counts)

# Create a new lightcurve object with normalized counts
normalized_lc = Lightcurve(lc.time, normalized_counts)

# Plot the normalized lightcurve
normalized_lc.plot()
plt.xlabel("Time (s)")
plt.ylabel("Normalized Counts")
plt.title("Normalized Lightcurve")
plt.show()
```
